```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from bs4 import BeautifulSoup
from selenium import webdriver

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
#https://public.opendatasoft.com/explore/dataset/us-zip-code-latitude-and-longitude/table/?refine.state=IL
zip_codes_file = 'us-zip-code-latitude-and-longitude-all.csv'

#https://data.cityofchicago.org/Health-Human-Services/COVID-19-Cases-Tests-and-Deaths-by-ZIP-Code/yhhz-zm2v
covid_file = 'COVID-19_Cases__Tests__and_Deaths_by_ZIP_Code.csv'

zip_code_df = pd.read_csv(zip_codes_file, sep=';', usecols=['Zip', 'City', 'Latitude', 'Longitude'])
covid_df = pd.read_csv(covid_file)

zip_code_df[zip_code_df.Zip == 46406]
```

```python
# Get all zip code from covid file
covid_zip_codes_df = DataFrame(np.sort(covid_df['ZIP Code'].replace('Unknown',np.nan).dropna().unique()), columns=['Zip'])
covid_zip_codes_df['Zip'] = covid_zip_codes_df['Zip'].astype('int64')

covid_zip_codes_df.head(3)
zip_code_df.head(3)
```

```python
# get all zip code which are present in both covid and zipcode files
joined_zip_codes_df = zip_code_df.merge(covid_zip_codes_df, how='inner', on='Zip').sort_values(by='Zip')
joined_zip_codes_df.head(3)
```

```python

# https://sites.google.com/a/chromium.org/chromedriver/home
    

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')

driver = webdriver.Chrome(executable_path='./chromedriver',options=chrome_options)
file = open("vzw_stores_raw.csv","w")
file.write("street,city,zip_code\n")

def save_stores_data(lat, lon, zip_code):
    url = "https://www.verizonwireless.com/stores/storesearchresults/?lat={}&long={}&q={}#/Search".format(lat,lon,zip_code)
    # url = 'https://www.verizonwireless.com/stores/storesearchresults/?lat=41.8817767&long=-87.6371461&q=60606#/Search'
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    stores_p = soup.find_all(lambda tag: tag.name == 'p' and tag.get('class') == ['noMargin', 'col-xs-8', 'lineHeight1_5'])

    for s in stores_p:
        spans = s.find_all('span')
        street = str(s.contents[0]).title()
        city = str(spans[0].contents[0].split(",")[0]).title()
        zip_code = str(spans[1].contents[0].split(" ")[1])
        rec = "{},{},{}\n".format(street, city, zip_code)
        file.write(rec)

for index, row in joined_zip_codes_df.iterrows():
    save_stores_data(row['Latitude'], row['Longitude'], row['Zip'])
    print("Saved data of ==> ", row['Zip'], row['Latitude'], row['Longitude'])

file.flush()
file.close()
```

```python
vzw_stores_df = pd.read_csv('vzw_stores_raw.csv')
vzw_stores_df.head(5)
vzw_stores_df.info()
```

```python
vzw_stores_df.drop_duplicates(inplace=True)
vzw_stores_df.info()
```

```python
vzw_stores_df = vzw_stores_df.merge(zip_code_df, how='left', left_on='zip_code', right_on='Zip')
vzw_stores_df = vzw_stores_df[['street', 'city', 'zip_code', 'City', 'Latitude', 'Longitude']]
```

```python
vzw_stores_df.to_csv('vzw_stores_final.csv')
```
